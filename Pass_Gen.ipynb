```python
# import radom module
import random
```

```python
# User input for password length
pass_length = int(input("Enter the length of password :  "))

#Storing letters,numbers & special characters 
special_chrs = "abcdefghijklmnopqrstuvwxyz0123456789ABCDFGHIJKLMNOPQRSTUVWXYZ!@#$%^&*()?"


#random sampling by joining the length of the password and the varible special_chrs
test = "".join(random.sample(special_chrs,pass_length))
print(test)
```
